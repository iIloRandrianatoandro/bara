```python
from transformers import VitsModel, AutoTokenizer
import torch
import scipy.io.wavfile

model = VitsModel.from_pretrained("facebook/mms-tts-mlg")
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-mlg")

model.eval()
text = "Miarahaba anareo rehetra amin'ity andro manokana ity"

inputs = tokenizer(text, return_tensors="pt")
with torch.no_grad():
    waveform = model(**inputs).waveform[0].numpy()

scipy.io.wavfile.write("tts_malagasy.wav", rate=model.config.sampling_rate, data=waveform)
```

```python
#Recharger le modèle et le tokenizer depuis Drive
from transformers import VitsModel, AutoTokenizer
import torch
import scipy.io.wavfile

# Chemin vers le dossier contenant le modèle sauvegardé
model_path = "/content/drive/MyDrive/projet_any_ampianarana/bara/tts_mg_officiel"

# Recharger
model = VitsModel.from_pretrained(model_path, local_files_only=True)
tokenizer = AutoTokenizer.from_pretrained(model_path, local_files_only=True)
```

```python
text = "handeha hianatsy iaho"

# Tokeniser
inputs = tokenizer(text, return_tensors="pt")

# Générer l'audio
with torch.no_grad():
    waveform = model(**inputs).waveform[0].numpy()

# Enregistrer dans un fichier WAV (dans Drive ou local)
output_path = "/content/drive/MyDrive/projet_any_ampianarana/bara/tts_malagasy.wav"
scipy.io.wavfile.write(output_path, rate=model.config.sampling_rate, data=waveform)

print(f"✅ Fichier audio enregistré à : {output_path}")
```
